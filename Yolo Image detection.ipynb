```python
import torch

# Load the YOLOv5 model (pre-trained on COCO dataset)
model = torch.hub.load('ultralytics/yolov5', 'yolov5s', pretrained=True)
```

```python
import cv2
import matplotlib.pyplot as plt

# Load image using OpenCV
img_path = 'dog_bike_car.jpg'
img = cv2.imread(img_path)

# Convert the image from BGR to RGB (for correct visualization)
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

# Display the image using Matplotlib
plt.imshow(img_rgb)
plt.axis('off')  # Hide axis
plt.show()
```

```python
# Perform inference on the image
results = model(img_rgb)

# Show detection results
results.show()  # This will display the image with bounding boxes
```

```python
# Results are in a DataFrame
df_results = results.pandas().xyxy[0]  # Bounding boxes in (xmin, ymin, xmax, ymax) format

# Print detection results
print(df_results)
```
